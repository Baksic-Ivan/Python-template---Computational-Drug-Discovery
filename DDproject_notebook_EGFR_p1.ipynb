```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## Searching for Epidermal growth factor receptor (EGFR) target protein and filtering 'SINGLE PROTEIN' target type

```python
target = new_client.target
target_query = target.search('EGFR')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
targets_filtered = targets.query("target_type == 'SINGLE PROTEIN'")
targets_filtered
```

I selected human EGFR as my target protein

```python
my_target = targets_filtered.target_chembl_id[3]
my_target 
```

## Retrieving bioactivity data for EGFR (this takes around 10 minutes)


Here IC50 data represents drug concentration necessary for 50% of receptor inhibition. This means that drug with lower IC50 has more potency.

```python
activity = new_client.activity
EGFR_activity = activity.filter(target_chembl_id = my_target).filter(standard_type = "IC50")

activity_df = pd.DataFrame.from_dict(EGFR_activity)
activity_df
```

Saving raw table

```python
activity_df.to_csv('EGFR_bioactivity_data_raw.csv', index=False)
```

Exclude missing data and duplicates

```python
activity_df_filtered = activity_df[activity_df.standard_value.notna() & activity_df.canonical_smiles.notna()]
activity_df_filtered
```

```python
activity_df_filtered_nr = activity_df_filtered.drop_duplicates(["canonical_smiles"])
activity_df_filtered_nr
```

Filtering columns and labeling compounds as either being active, inactive or intermediate based on their IC50 value

```python
selection = ["molecule_chembl_id", "canonical_smiles", "standard_value"]
activity_df_filtered_nr_selection = activity_df_filtered_nr[selection].reset_index(drop = True)
activity_df_filtered_nr_selection

```

```python
activity_df_filtered_nr_selection.to_csv("EGFR_bioactivity_data_preprocessed.csv", index = False)
```

```python
bioactivity_threshold = []
for i in activity_df_filtered_nr_selection.standard_value:
    if float(i) >= 10000:
        bioactivity_threshold.append("inactive")
    elif float (i) <= 1000:
        bioactivity_threshold.append("active")
    else:
        bioactivity_threshold.append("intermediate")

bioactivity_class = pd.Series(bioactivity_threshold, name = "class")
activity_df_filtered_nr_selection_class = pd.concat([activity_df_filtered_nr_selection, bioactivity_class], axis = 1)
activity_df_filtered_nr_selection_class
```

```python
activity_df_filtered_nr_selection_class.to_csv("EGFR_bioactivity_data_curated.csv", index = False)
```
